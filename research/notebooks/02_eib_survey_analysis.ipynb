# EIB Investment Survey — Trade Disruption Obstacles in Lithuania

**Dataset:** European Investment Bank (EIBIS) survey data, Lithuania (2018–2025)

**Research Question:** What trade-related obstacles do Lithuanian firms face? How does this vary by sector and company size?

**Source:** [EIB Investment Survey](https://data.eib.org/eibis)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/eib_survey_lithuania.csv')
print(f'Rows: {len(df)}')
print(f'Years: {sorted(df["Survey wave"].unique())}')
df.head(10)
```

## Trade Disruption Trends
The share of Lithuanian firms reporting trade disruptions as an obstacle has grown significantly since 2023, peaking at 36% in 2023 before moderating to 25% in 2024.

## Key Findings

1. **25% of Lithuanian firms** reported trade disruptions as an obstacle in 2024
2. **Manufacturing** is the most affected sector (up to 11% reporting it as a major obstacle)
3. **Medium/Large firms** report higher disruption rates than micro/small — but small firms lack resources to cope
4. Trade obstacles **spiked in 2023** (post-COVID supply chain disruptions + geopolitical tensions)

## Implications for XaaS Platform

- Strong and growing demand signal — trade disruptions are a real, measured problem
- Manufacturing exporters are the most affected — aligns with Comtrade data showing machinery as top export
- Small firms need the most help but have fewest resources — supports a SaaS/platform pricing model
- Timing is right: obstacle awareness is high, firms are actively seeking solutions
